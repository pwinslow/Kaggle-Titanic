```python
# Import division from future
from __future__ import division

# Basic analysis imports
import pandas as pd
from pandas import Series,DataFrame
from pandas.tools import plotting
import numpy as np

# Scikitlearn imports
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import StratifiedKFold
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.metrics import roc_curve, auc

# Standard visualization analysis imports
import matplotlib.pyplot as plt
import seaborn as sb
sb.set_style('whitegrid')
%matplotlib inline
```

```python
# Read in data and look at basic info 
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')
train_df.drop('PassengerId', axis=1, inplace=True)

'''VARIABLE DESCRIPTIONS:
survival        Survival
                (0 = No; 1 = Yes)
pclass          Passenger Class
                (1 = 1st; 2 = 2nd; 3 = 3rd)
name            Name
sex             Sex
age             Age
sibsp           Number of Siblings/Spouses Aboard
parch           Number of Parents/Children Aboard
ticket          Ticket Number
fare            Passenger Fare
cabin           Cabin
embarked        Port of Embarkation
                (C = Cherbourg; Q = Queenstown; S = Southampton)

SPECIAL NOTES:
Pclass is a proxy for socio-economic status (SES)
 1st ~ Upper; 2nd ~ Middle; 3rd ~ Lower

Age is in Years; Fractional if Age less than One (1)
 If the Age is Estimated, it is in the form xx.5

With respect to the family relation variables (i.e. sibsp and parch)
some relations were ignored.  The following are the definitions used
for sibsp and parch.

Sibling:  Brother, Sister, Stepbrother, or Stepsister of Passenger Aboard Titanic
Spouse:   Husband or Wife of Passenger Aboard Titanic (Mistresses and Fiances Ignored)
Parent:   Mother or Father of Passenger Aboard Titanic
Child:    Son, Daughter, Stepson, or Stepdaughter of Passenger Aboard Titanic

Other family relatives excluded from this study include cousins,
nephews/nieces, aunts/uncles, and in-laws.  Some children travelled
only with a nanny, therefore parch=0 for them.  As well, some
travelled with very close friends or neighbors in a village, however,
the definitions do not support such relations.
'''
print 'Training set info...'
print train_df.info()
print '\nTesting set info...'
print test_df.info()
```

```python
# Define a number of methods to fill in missing values based on the corresponding median/mean of all similar instances
def age_filler(df):
    
    df = df.copy()
    noage = df[ df.Age.isnull() ]
    if noage.shape[0] == 0:
        print 'No null age values...'
        return df.Age
    else:
        for row in range(len(noage)):
            temp_ser = noage.iloc[row]
            temp_df = df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex) 
                                & (df.SibSp == temp_ser.SibSp) 
                                & (df.Parch == temp_ser.Parch) 
                                & (df.Age.notnull()) ]
            if temp_df.shape[0] > 0:
                mean_age = round( temp_df.Age.mean() )
                df.loc[temp_ser.name, 'Age'] = mean_age
            elif df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex) 
                                & (df.Parch == temp_ser.Parch)
                                & (df.Age.notnull()) ].shape[0]>0:
                temp_df = df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex) 
                                & (df.Parch == temp_ser.Parch)
                                & (df.Age.notnull()) ]
                mean_age = round( temp_df.Age.mean() )
                df.loc[temp_ser.name, 'Age'] = mean_age
            elif df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex) 
                                & (df.Age.notnull()) ].shape[0]>0:
                temp_df = df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex) 
                                & (df.Age.notnull()) ]
                mean_age = round( temp_df.Age.mean() )
                df.loc[temp_ser.name, 'Age'] = mean_age

        # Check to see that all nulls were replaced
        null_num = df[ df.Age.isnull() ].shape[0]
        if null_num == 0:
            print 'All null age values successfully replaced...'
        else:
            print 'Whoops, missed some...'

        return df.Age

def deck_filler(df):
    
    df = df.copy()
    nodeck = df[ df.Deck.isnull() ]
    if nodeck.shape[0] == 0:
        print 'No null deck values...'
        return df.Deck
    else:
        for row in range(len(nodeck)):
            temp_ser = nodeck.iloc[row]
            temp_df = df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex) 
                                & df.Deck.notnull() ]
            if temp_df.shape[0] > 0:
                median_deck = temp_df.Deck.median()
                df.loc[temp_ser.name, 'Deck'] = median_deck
            elif df[ (df.Pclass == temp_ser.Pclass)  
                                & df.Deck.notnull() ].shape[0]>0:
                temp_df = df[ (df.Pclass == temp_ser.Pclass)  
                                & df.Deck.notnull() ]
                median_deck = temp_df.Deck.median()
                df.loc[temp_ser.name, 'Deck'] = median_deck

        # Check to see that all nulls were replaced
        null_num = df[ df.Deck.isnull() ].shape[0]
        if null_num == 0:
            print 'All null cabin (i.e., deck) values successfully replaced...'
        else:
            print 'Whoops, missed some...'

        return df.Deck

def embarkation_filler(df):
    
    df = df.copy()
    noemb = df[ df.Embarked.isnull() ]
    if noemb.shape[0] == 0:
        print 'No null embarked values...'
        return df.Embarked
    else:
        for row in range(len(noemb)):
            temp_ser = noemb.iloc[row]
            temp_df = df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex)
                                & (df.Deck == temp_ser.Deck) 
                                & df.Embarked.notnull() ]
            if temp_df.shape[0] > 0:
                median_embarked = temp_df.Embarked.median()
                df.loc[temp_ser.name, 'Embarked'] = median_embarked

        # Check to see that all nulls were replaced
        null_num = df[ df.Embarked.isnull() ].shape[0]
        if null_num == 0:
            print 'All null embarked values successfully replaced...'
        else:
            print 'Whoops, missed some...'

        return df.Embarked

def fare_filler(df):
    
    df = df.copy()
    nofare = df[ df.Fare.isnull() ]
    if nofare.shape[0] == 0:
        print 'No null fare values...'
        return df.Fare
    else:
        for row in range(len(nofare)):
            temp_ser = nofare.iloc[row]
            temp_df = df[ (df.Pclass == temp_ser.Pclass) 
                                & (df.Sex == temp_ser.Sex)
                                & (df.Deck == temp_ser.Deck) 
                                & (df.Embarked == temp_ser.Embarked)
                                & (df.Fare.notnull()) ]
            if temp_df.shape[0] > 0:
                mean_fare = temp_df.Fare.mean()
                df.loc[temp_ser.name, 'Fare'] = mean_fare

        # Check to see that all nulls were replaced
        null_num = df[ df.Fare.isnull() ].shape[0]
        if null_num == 0:
            print 'All null fare values successfully replaced...'
        else:
            print 'Whoops, missed some...'

        return df.Fare
    
# Define a method to perform some basic cleaning and run all previous methods to fill in nans
def clean_data(df):
    
    print 'Starting with %d null values' % df.isnull().values.sum()
    df.Age = age_filler(df)
    deck_map = {'A':0, 'B':1, 'C':2, 'D':3, 'E':4, 'F':5, 'G':6, 'T':7}
    df['Deck'] = df.Cabin.map(lambda x: deck_map[x[0]] if pd.notnull(x) else x)
    df.Deck = deck_filler(df)
    df.Embarked = df.Embarked.map({'S':0, 'C':1, 'Q':2})
    df.Embarked = embarkation_filler(df)
    df.Fare = fare_filler(df)
    df.Sex = df.Sex.map({'male':0, 'female':1})
    df['Family'] = (df.SibSp + df.Parch).map(lambda x: 1 if x > 0 else 0)
    title_list = list(np.unique([name.split(', ')[1].split('.')[0] for name in df.Name]))
    title_map = {}
    for cnt, title in enumerate(title_list):
        title_map[title] = cnt
    df['Title'] = df.Name.apply(lambda x: title_map[x.split(', ')[1].split('.')[0]])
    df.drop(['Cabin', 'Ticket', 'Name', 'SibSp', 'Parch'], axis=1, inplace=True)
    print 'Ending with %d null values\n' % df.isnull().values.sum()
    
    return (df, title_map)

# Clean the train and test datasets
print '\nCleaning data...'
train_df, title_map = clean_data(train_df)
test_df, _ = clean_data(test_df)
```

```python
# As might be expected, your chance of survival goes down with your social standing
sb.countplot(x='Pclass', hue='Survived', data=train_df)
plt.title('Class and Survival (0:Died, 1:Survived)')
plt.show()
# Females had a much greater chance of survival than Males
sb.countplot(x='Sex', hue='Survived', data=train_df)
plt.title('Sex (0:Male, 1:Female) and Survival')
plt.show()
# Fare amount positively affects survival probability but only if you're in 1st. Otherwise, it doesn't seem to make
# much of a difference.
sb.factorplot(x='Survived', y='Fare', hue='Pclass', data=train_df)
plt.title('Fare vs Survival and Class')
plt.show()
# In all classes, the young are more likely to live
sb.factorplot(x='Survived', y='Age', hue='Pclass', data=train_df)
plt.title('Age vs Survival and Class')
plt.show()
# Peoples embarkation is a relatively good indicator of survival... Maybe this is simply a stand-in for class?
sb.countplot(x='Embarked', hue='Survived', data=train_df)
plt.title('Embarked (0:Southampton, 1:Cherbourg, 2:Queenstown) and Survival')
plt.show()
# Certain decks had much better chances of survival than others. Perhaps this speaks to where the lifeboats were 
# located?
sb.countplot(x='Deck', hue='Survived', data=train_df)
plt.title('Deck and Survival')
plt.show()
# Those with families were less likely to die
sb.countplot(x='Family', hue='Survived', data=train_df)
plt.title('Family (0:No Family, 1:Had Family) and Survival')
plt.show()
# Better to be a "Miss"(8) or "Mrs"(12) but definitely don't want to be a "Mr"(11)
sb.countplot(x='Title', hue='Survived', data=train_df)
plt.title('Title (e.g., Mr., Mrs., Master., etc.) and Survival')
plt.show()
```

```python
print 'Beginning 5-fold cross validation grid search over random forest model parameters...'

# Create numpy arrays to train/test a random forest classifier
X = train_df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'Deck', 'Family', 'Title']].values
Y = train_df.Survived.values
X_eval = test_df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'Deck', 'Family', 'Title']].values

# Split the development data into a training and a testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, 
                                                    random_state=492, stratify=Y)

rand_forest = RandomForestClassifier()

# Define a parameter grid to search over
param_grid = {"n_estimators": range(5,55,5), "max_depth": range(1,11,1), 
              'criterion':['gini','entropy']}

# Perform 5-fold stratified cross validation
cross_validation = StratifiedKFold(Y_train, n_folds=5)

# Perform grid search
clf = GridSearchCV(rand_forest, param_grid=param_grid, cv=cross_validation, 
                                    scoring='roc_auc', n_jobs=10, verbose=10)
_ = clf.fit(X_train, Y_train)
```

```python
# Generate ROC curve and calculate AUC score on test set
print 'Using best estimator from grid search to generate ROC curve and AUC score on test set...'

Y_predicted = clf.best_estimator_.predict(X_test)
print classification_report(Y_test, Y_predicted,
                            target_names=["Died", "Survived"])
print "Area under ROC curve: {:0.5f}".format(roc_auc_score(Y_test,
                                                  clf.best_estimator_.predict_proba(X_test)[:,1]))

decisions = clf.best_estimator_.predict_proba(X_test)[:,1]
# Compute ROC curve and area under the curve
fpr, tpr, thresholds = roc_curve(Y_test, decisions)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.plot(fpr, tpr, lw=1, label='ROC (area = %0.3f)'%(roc_auc))
plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) curve')
plt.legend(loc="lower right", frameon = True).get_frame().set_edgecolor('black')
plt.grid(True, linestyle = 'dotted')
plt.show()
```

```python
importances = clf.best_estimator_.feature_importances_
indices = np.argsort(importances)[::-1]
Feature_List = list(train_df.columns)
Feature_List.remove('Survived')

# Plot the feature importances of the bdt
plt.figure(figsize=(15,10))
plt.title("Feature importances", fontsize=20)
plt.barh(range(X_eval.shape[1]), importances[indices],
       color="salmon", align="center")
plt.yticks(range(X_eval.shape[1]), [Feature_List[i] for i in indices], fontsize=15)
plt.ylim([-1, X_eval.shape[1]])
plt.show()
```

```python
# Create kaggle submission csv for evaluation set
DataFrame({'PassengerId':test_df.PassengerId, 
           'Survived':clf.best_estimator_.predict(X_eval)}).to_csv('../data/titanic_submission.csv', index=False)
```
